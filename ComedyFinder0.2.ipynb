```python
import numpy as np
import csv

def nonlin(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))

with open('train.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    next(csvfile)
    
    LisTarget = [ ]
    
    LisAction = [ ]
    LisCrime = [ ]
    LisBiography = [ ]
    LisFamily = [ ]
    LisHorror = [ ]
    LisMusic = [ ]
    LisMusical = [ ]
    LisThriller = [ ]
    LisWar = [ ]
    LisWestern = [ ]
    LisDrama = [ ] 
    
    for row in readCSV:
        isAction = int(row[6])
        isCrime = int(row[10])
        isBiography = int(row[9])
        isFamily = int(row[13])
        isHorror = int(row[16])
        isMusic = int(row[17])
        isMusical = int(row[18])
        isThriller = int(row[24])
        isWar = int(row[25])
        isWestern = int(row[26])
        isDrama = int(row[12])
        
        isTarget = int(row[1])
        
        LisAction.append(isAction)
        LisCrime.append(isCrime)
        LisBiography.append(isBiography)
        LisFamily.append(isFamily)
        LisHorror.append(isHorror)
        LisMusic.append(isMusic)
        LisMusical.append(isMusical)
        LisThriller.append(isThriller)
        LisWar.append(isWar)
        LisWestern.append(isWestern)
        LisDrama.append(isDrama)
        
        LisTarget.append(isTarget)
        
syn0 = 2*np.random.random((2,1)) - 1
syn1 = 2*np.random.random((5,1)) - 1

for x in range(1, len(LisTarget)):
    X = [[LisAction[x], LisCrime[x]], 
         [LisBiography[x], LisFamily[x]],
         [LisHorror[x], LisMusic[x]], 
         [LisMusical[x], LisThriller[x]], 
         [LisWar[x], LisWestern[x]]]

    Y = LisTarget[x]

    l0=X
    l1 = nonlin(np.dot(l0, syn0))
    l2 = nonlin(np.dot(l1.T, syn1))


    l2error = Y - l2
    l2delta = l2error*nonlin(l2,deriv=True)

    l1error = l2delta.dot(syn1.T)

    l1delta = l1error * nonlin(l1,deriv=True)


    syn1 += l1.dot(l2delta)
    
    print(x)
    print(l2error)
    
    
```
